## Logistic Regression

```python
# Split the data
y = df.diagnosis
X = df.drop('diagnosis',axis=1)

```

```python
X
```

```python
#Data normalization which reduce values close to values of 0,1 and -1
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X))

```

```python
X_scaled
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=103)
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(X_train,y_train)
lr_pred = lr.predict(X_train)
lr_pred2 = lr.predict(X_test)
```

```python
lr_pred
```

```python
lr_pred2
```

```python
#Model Evaluation
from sklearn.metrics import accuracy_score,classification_report

accuracy = accuracy_score(y_train,lr_pred)
accuracy
```

```python
#Classification report
print(f'Accuracy:{accuracy:.2f}')
print(classification_report(y_train,lr_pred))
```

```python
#Model Evaluation
from sklearn.metrics import accuracy_score,classification_report

accuracy = accuracy_score(y_test,lr_pred2)
```

```python
accuracy
```

```python
#Classification report
print(f'Accuracy:{accuracy:.2f}')
report = pd.DataFrame(classification_report(y_test,lr_pred2))
report
```
